````python
### 수정했음!!444
!pip3 install requests
!pip3 install bs4
import requests
from typing import List
from bs4 import BeautifulSoup  # html 파싱을 위한 라이브러리
import re # 정규표현식 사용을 위한 라이브러리


# 다음의 웹페이지에서, 상품에 해당하는 이미지의 고유 아이디를 파싱해주세요!
URL = "http://www.pythonscraping.com/pages/page3.html"
````

````python
# pattern의 활용예시
corpus = [
          "123 adsfdsf",
          "afafff 123"
]
pattern = re.compile(r'^123')

for doc in corpus:
  match = pattern.match(doc)
  print(match)
````

````python
# requests 라이브러리를 활용하면 한줄로 html을 간단하게 로드할 수 있습니다!
html = requests.get(URL).text
soup = BeautifulSoup(html, "html.parser")
````

````python
# 다시 말하지만, "상품"에 해당하는 이미지의 id만 가져오면 됩니다~
for tag in soup.find_all('img'):
  print(tag)
````

````python
# 1-1. 정규식으로 img태그 가져오기
imgs1 = soup.find_all('img', {'src' : re.compile('img(\d+).jpg')})

# 1-2. img태그에서 src속성 가져오기
img_ids1 = [
            tag['src']
            for tag in imgs
]

# 1-3. src속성을 '/'로 분리해서 맨 마지막에 해당하는 정보 가져오기
img_ids1_1 = [
              url.split('/')[-1]
              for url in img_ids1
]

print(imgs1)
print(img_ids1)
print(img_ids1_1)
````

````python
# 2-1. css 선택자로 img태그 가져오기
imgs2 = soup.select('tr.gift > td > img')

# 2-2. src속성에서 정규식으로 가져오기
img_ids2 = [
            re.search('img\d+\.jpg', img['src']).group(0)
            for img in imgs2
]

print(imgs2)
print(img_ids2)
````
